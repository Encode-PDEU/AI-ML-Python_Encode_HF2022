```python
# Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
# Loading Dataset

data = pd.read_csv("store_data.csv" , header = None)
display(data.head())
print(data.shape)
```

```python
# Data Preprocessing

rec = []
for i in range(1, 7501):
    rec.append([str(data.values[i, j]) for j in range(0, 20)])

print(type(rec))
```

```python
# Implementation of Algorithm

association_rules = apriori(rec, min_support=0.0042, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
# Total rules derived

for i in range(0, len(association_results)):
    print(association_results[i][0])
```

```python
# Visualization of rules generated

for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " > " + items[1])

    print("Support: " + str(item[1]))

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print(end="/n")
```
