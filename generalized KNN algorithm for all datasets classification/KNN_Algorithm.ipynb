```python
import pandas as pd
import numpy as np

df = pd.read_csv("diabetes.csv")
```

```python
df
total_cols=len(df.axes[1])
print(total_cols)
```

```python
print("Enter the values below...")
```

```python
random_arr = [0 for _ in range(total_cols-1)]
for i in range(len(random_arr)):
    inp = float(input())
    random_arr[i] = inp
    
random_arr = np.array(random_arr)
random_arr
```

```python
print("Enter value of K below...")
```

```python
k = int(input())
```

```python
for ind, val in enumerate(random_arr):
    newdf = np.empty(shape=df.shape[0])
    for i in range(len(newdf)):
        newdf[i] = (df.iat[i,ind] - val)**2
    tempstr = f"temp{ind}"
    df[tempstr] = newdf

df["total"] = df.temp0 + df.temp1 + df.temp2 + df.temp3
df["sqrt"] = np.sqrt(df["total"])
```

```python
df
```

```python
df = df.sort_values("sqrt", ascending=True)
df
```
